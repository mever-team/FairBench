# Computing a fairness report

```python
predictions = [1, 1, 0, 1, 0, 0, 1, 0, 0]
labels = [1, 0, 0, 1, 0, 1, 0, 1, 0]
sensitive = ["M", "F", "M", "F", "M", "F", "O", "O", "M"]
```

```python
sensitive = fb.Dimensions(fb.categories@sensitive)
sensitive
```

```python
sensitive["F"].numpy()
```

```python
acc = fb.blocks.measures.acc(predictions=predictions, labels=labels, sensitive=sensitive["F"].numpy())
float(acc)
```

```python
report = fb.reports.vsall(predictions=predictions, labels=labels, sensitive=sensitive, measures=[fb.measures.acc, fb.measures.pr])
report.show(env=fb.export.Console(ansiplot=False), depth=2)
```

```python
report = fb.reports.vsall(predictions=predictions, labels=labels, sensitive=sensitive, measures=[fb.measures.acc, fb.measures.pr])
html = report.show(env=fb.export.Html(filename=None), depth=2)
HTML(html)
```

```python
report.acc.min.samples.show()
```

# Creating and exploring a full report

```python
predictions = [1, 1, 0, 1, 0, 0, 1, 0, 0]
labels = [1, 0, 0, 1, 0, 1, 0, 1, 0]
sensitive = ["M", "F", "M", "F", "M", "F", "O", "O", "M"]
sensitive = fb.Dimensions(fb.categories@sensitive)
report = fb.reports.vsall(predictions=predictions, labels=labels, sensitive=sensitive)
report.show()
```

```python
report.help()
```

```python
report.show(env=fb.export.ConsoleTable)
```

```python
report.show(env=fb.export.ConsoleTable(sideways=False), depth=2)
```
